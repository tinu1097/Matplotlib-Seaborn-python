# style lines and markers
the plt.plot function supports many arguments for styling lines and markers
colors or c: set color of the line 
linestyle or ls: choose b/w a solid or dashed line
linewidth or lw: set the width of a line
markersize or ms : set the size of marker
markeredgecolor or mec:set the edge color for markers
markerfacecolor or mfc:set the fill color for markers
alpha = opacity of the plot

```python
plt.plot(years,apples,marker='s',c='b',ls='--',lw=3,ms=5,mew=3,mec='navy')
plt.plot(years,oranges,marker='o',c='r',ls='-',lw=5,ms=10,alpha=0.5) 
plt.plot(years,oranges,marker = 'x')
plt.xlabel('YEAR')
plt.ylabel('YIELD(TONS PER HECTURES)')
plt.title("CROP YIELDS IN KANTO")
plt.legend(["APPLES","ORANGES"]);
```

# Scatter Plot(bivariate&multivariate)
A scatter plot, or scatterplot, is generally used to summarize the relationship between two paired data samples. Paired data samples means that two measures were recorded for a given observation, such as the weight and height of a person. The x-axis represents observation values for the first sample, and the y-axis represents the observation values for the second sample. Each point on the plot represents a single observation.They help us understand relationships among the variables of a dataset.
scatter plot visualize relationship between two varaibles.
in a scatter plot,the variables of 2 variables are plotted as points on a 2-dimensional grid.
additionally , we can also use a third variable to determine the size or color of the points.

is statistically termed as their correlation. As we’ve previously observed, closer the data points come when plotted to making a straight line, higher the correlation between those two variables, thus making their relationship stronger.

# Histogram
A histogram plot is generally used to summarize the distribution of a data sample. The x-axis
represents discrete bins or intervals for the observations and  y-axis represents the frequency or count of the number of observations in the dataset that belong to each bin. Essentially, a data sample is transformed into a bar chart where each category on the x-axis represents an interval of observation values.The histogram can be classified based on binwidth as unimodal, bimodal, left skewed, right skewed and symmetric skewed plots.

A histogram represents the distribution of data by forming bins along the range of the data and then drawing bars to show the number of observations that fall in the bin.

# Plotting Multiple Charts in a grid 
Matplotlib and Seaborn also support plotting multiple charts in a grid ,using plt.subplots which returns a set of axes that can be used for plotting.

```python
# axis total number of values (0,0) its measns 0th row,0th column
fig,axes = plt.subplots(2,3,figsize=(13,10));
axes[0,0].plot(years,oranges,'o--r')
axes[0,0].plot(years,apples,'s-b')
axes[0,0].set_title("crops production")
axes[0,0].set_xlabel("year")
axes[0,0].set_ylabel("yield(ton hec)")
axes[0,0].legend(["oranges","apples"])

#scatterplot
axes[0,1].set_title('sepal length vs.width')
sns.scatterplot(flow_df.sepal_length,
                flow_df.sepal_width,
                hue=flow_df.species,
                s=80,
               ax=axes[0,1]);

#histogram 
axes[0,2].set_title("histogram for sepal_width")
sns.histplot(flow_df.sepal_width,
            ax = axes[0,2])

#barplot
axes[1,0].set_title("bar plot for female & male")
sns.barplot(tips_df.total_bill,
            tips_df.day,
           hue = tips_df.sex,
           ax = axes[1,0])
#heatmap
axes[1,1].set_title("heatmap for pessengers")
sns.heatmap(flights_df1,
           ax = axes[1,1])
#advance heatmap
axes[1,2].set_title("advance heatmap for pessengers")
sns.heatmap(flights_df1,
           fmt = "d",
           annot = True,
           cmap="Blues",
           ax = axes[1,2])
plt.tight_layout(pad=3)  #for making some spaces b/w  plots 
```
