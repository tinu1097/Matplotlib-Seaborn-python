# HISTOGRAM
A histogram plot is generally used to summarize the distribution of a data sample. The x-axis
represents discrete bins or intervals for the observations and  y-axis represents the frequency or count of the number of observations in the dataset that belong to each bin. Essentially, a data sample is transformed into a bar chart where each category on the x-axis represents an interval of observation values.The histogram can be classified based on binwidth as unimodal, bimodal, left skewed, right skewed and symmetric skewed plots.

A histogram represents the distribution of data by forming bins along the range of the data and then drawing bars to show the number of observations that fall in the bin. 

```python
x = np.random.randn(100)
sns.distplot(x);
```

#defining cmap can change the colours to our desired colours.
#linewidth is the gape between each squares
#setting annot as True will show the correlation values in the heatmap

```python
#CORRALTION IS ALWAYS APPLICABEL ON INT AND FLOAT DATATYPE 
#CORRELATION RANGE IS -1 TO +1 
#-1 MEANS HIGHLY NEAGTIVE CORR AND +1 IS HIGHLY POSITIVE CORR
# AND 0 MEANS NO CORRELATION
```

Plots on Categorical Data
We are going to see about the various plots on categorical data. There are three types of categorical data representation. They are as follow:
· Show each observation- In this type each and every observation is plotted as distinct points. It comprises the following plots:-
1)Strip Plots
2)Swarm Plots
· Abstract representation- These plots show the abstract view of the observations. The plots come under this category are as follows:-
1)Box plots
2)Violin plots
· Statistical Estimates- This type of representation shows the plot along with statistical information such as confidence interval levels.
1)Bar Plot
2)Point Plot
3)Count Plot
Let us see about each and every type of categorical representations.

#BOXPLOT
A box and whisker plot, or boxplot for short, is generally used to summarize the distribution of a data sample. The x-axis is used to represent the data sample, where multiple boxplots can be drawn side by side on the x-axis if desired. The y-axis represents the observation values. A box is drawn to summarize the middle 50% of the dataset starting at the observation at the 25th percentile and ending at the 75th percentile. The median, or 50th percentile, is drawn with a line.

```python
sns.boxplot(data=df,orient="v")
#orient v - vertival diagram
```

```python
plt.figure(figsize=(5,5))
sns.boxplot(x="total_bill",y="day",data=df,orient="h");
```

```python
sns.boxplot(x="day",y="total_bill",hue="smoker",data=df,palette="Set3")
```

The violin plot is the combination of kernel density estimates and box plot. Kernel Density Estimation (KDE) is a way to estimate the probability density function of a continuous random variable. It is used for non-parametric analysis. Hence it is easier to analyze and understand the distribution of the data.

```python
#VIOLIN IS SIMILAR TO BOXPLOT BUT INSTEAD REPRESENATION IS IN FORM OF VIOLIN (OR) GUITAR
#inside violin plot we have boxplot
sns.catplot(x="sex",y="total_bill",hue="sex",data=df,kind="violin");
```

```python
sns.catplot(x='total_bill',y='day',hue='sex',kind='violin',split=False,data=df,inner='stick')
```

```python
sns.catplot(x='total_bill',y='day',hue='sex',kind='violin',data=df,inner='stick');
```
